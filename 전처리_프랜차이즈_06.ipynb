## 프랜차이즈와 일반카페 데이터프레임 나누기

## 프랜차이즈 퀄리티별 종류

```python
franchise_up2 =[]
for i in franchise_up:
    franchise_up2.append(df2['사업장명'].str.contains(i))

franchise_up2
```

```python
df_li = pd.DataFrame(franchise_up2).T
df_li
```

```python
df_li.columns = franchise_up
```

```python
df_li
```

```python
for i in range(len(df_li.columns)):
    for j in range(len(df_li.index)):
        if df_li.iloc[j,i] == True:
            df_li.iloc[j,i] = 1
        else:
            df_li.iloc[j,i] = 0
df_li
```

```python
df_li['franchise'] = df_li.sum(axis=1)
df_li
```

```python
df2['franchise'] = df_li['franchise']
df2
```
